# Exploratory Data Analysis (EDA) - Speech Classification
## Understanding the Expresso Dataset
This notebook performs an in-depth exploratory data analysis (EDA) on the Expresso dataset.  
We'll visualize speaking styles, extract audio features, and analyze class distributions.

```python
import numpy as np
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
```

## Load Dataset
The Expresso dataset contains audio clips labeled with different speaking styles.  
We'll load the dataset and check some samples.

```python
ds = load_dataset("ylacombe/expresso")
df = pd.DataFrame(ds["train"])
df.head()
```

## Class Distribution
Let's visualize how many samples exist for each speaking style.

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df["style"], order=df["style"].value_counts().index, palette="viridis")
plt.title("Speaking Style Distribution", fontsize=14)
plt.xlabel("Count")
plt.ylabel("Speaking Style")
plt.show()
```

## Extracting Audio Features
To analyze the dataset, we'll extract key audio features like:
- **Spectral Centroid**: Indicates where the center of mass of the frequency spectrum is located.
- **Zero Crossing Rate (ZCR)**: Measures how frequently the signal changes sign.
- **Root Mean Square (RMS) Energy**: Represents the loudness of the signal.

```python
def plot_audio_features(audio_array, sr):
    plt.figure(figsize=(15, 5))

    # Plot Waveform
    plt.subplot(3, 1, 1)
    librosa.display.waveshow(audio_array, sr=sr)
    plt.title("Waveform")

    # Compute Spectrogram
    plt.subplot(3, 1, 2)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(audio_array)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis="time", y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title("Spectrogram")

    # Compute Mel Spectrogram
    plt.subplot(3, 1, 3)
    mel_spect = librosa.feature.melspectrogram(y=audio_array, sr=sr, n_mels=128)
    mel_spect_db = librosa.power_to_db(mel_spect, ref=np.max)
    librosa.display.specshow(mel_spect_db, sr=sr, x_axis="time", y_axis="mel")
    plt.colorbar(format="%+2.0f dB")
    plt.title("Mel Spectrogram")

    plt.tight_layout()
    plt.show()

# Select a random sample
sample = ds["train"][0]
audio_array = sample["audio"]["array"]
sr = sample["audio"]["sampling_rate"]

# Plot the features
plot_audio_features(audio_array, sr)
```

## Comparing Features Across Styles
We'll now compare the extracted features (centroid, ZCR, RMS) for different speaking styles.

```python
features_by_style = {"style": [], "duration": [], "mean_centroid": [], "mean_zcr": [], "mean_rms": []}

for sample in ds["train"]:
    audio = sample["audio"]["array"]
    sr = sample["audio"]["sampling_rate"]
    style = sample["style"]

    # Extract features
    centroid = librosa.feature.spectral_centroid(y=audio, sr=sr)[0].mean()
    zcr = librosa.feature.zero_crossing_rate(audio)[0].mean()
    rms = librosa.feature.rms(y=audio)[0].mean()

    # Store in dictionary
    features_by_style["style"].append(style)
    features_by_style["duration"].append(len(audio) / sr)
    features_by_style["mean_centroid"].append(centroid)
    features_by_style["mean_zcr"].append(zcr)
    features_by_style["mean_rms"].append(rms)

# Convert to DataFrame
df_features = pd.DataFrame(features_by_style)

# Plot Distributions
plt.figure(figsize=(15, 10))
for i, feature in enumerate(["duration", "mean_centroid", "mean_zcr", "mean_rms"], 1):
    plt.subplot(2, 2, i)
    sns.violinplot(data=df_features, x="style", y=feature)
    plt.xticks(rotation=45)
    plt.title(f"{feature.replace('_', ' ').title()} by Style")

plt.tight_layout()
plt.show()
```
